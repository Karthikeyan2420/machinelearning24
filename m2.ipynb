```python
import pandas as pd


df = pd.read_csv('adult.csv')

total_count = len(df)
count_gt_50k = len(df[df['income'] == '>50K'])

marginal_prob_gt_50k = count_gt_50k / total_count
print(f"Marginal Probability of income >50K: {marginal_prob_gt_50k:.4f}")

count_bachelors = len(df[df['education'] == 'Bachelors'])
count_bachelors_gt_50k = len(df[(df['education'] == 'Bachelors') & (df['income'] == '>50K')])

conditional_prob_gt_50k_bachelors = count_bachelors_gt_50k / count_bachelors
print(f"Conditional Probability of income >50K given Bachelors: {conditional_prob_gt_50k_bachelors:.4f}")

count_gt_50k_private = len(df[(df['income'] == '>50K') & (df['workclass'] == 'Private')])

joint_prob_gt_50k_private = count_gt_50k_private / total_count
print(f"Joint Probability of income >50K and working in Private sector: {joint_prob_gt_50k_private:.4f}")
```

```python
#Normal Distribution

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

data=pd.read_csv('adult.csv')

mean_age=data['age'].mean()
std_age=data['age'].std()
print(mean_age)
print(std_age)
```

```python
x=np.linspace(mean_age-3*std_age,mean_age+3*std_age,100)
y=stats.norm.pdf(x,mean_age,std_age)

plt.figure(figsize=(10,6))
sns.histplot(data['age'],bins=10,kde=True,stat='density',color='skyblue',
            label='Age Distribution')
plt.plot(x,y,'r-',label="Normal Distribution",linewidth=2)
plt.legend()
plt
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the dataset
data = pd.read_csv('employee_salaries.csv')



# Calculate mean and standard deviation of salaries
mean_salary = data['Salary'].mean()
std_salary = data['Salary'].std()

# Generate values for the normal distribution
x = np.linspace(mean_salary - 3*std_salary, mean_salary + 3*std_salary, 100)
y = stats.norm.pdf(x, mean_salary, std_salary)

# Plotting the salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Salary'], bins=10, kde=True, stat='density', color='skyblue', label='Salary Distribution')
plt.plot(x, y, 'r-', label='Normal Distribution', linewidth=2)
plt.title('Normal Distribution of Employee Salaries')
plt.xlabel('Salary')
plt.ylabel('Density')
plt.legend()
plt.show()

# Print statistics
print(f"\nMean Salary: ${mean_salary:.2f}")
print(f"Standard Deviation: ${std_salary:.2f}")
```

```python
#correlation Analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("adult.csv")
num_df=df.select_dtypes(include=['int64','float64'])

cmat=num_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(cmat,annot=True,fmt=".2f",cmap="coolwarm",square=True,linewidths=.5)
plt
```
