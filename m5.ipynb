```python
#Gini index
import pandas as pd
def gini_coefficient(val):
    sorval=sorted(val)
    n=len(val)
    sumval=sum(sorval)
    gini=0
    for i,val in enumerate(sorval):
        gini+=(i+1)*val
    giniden=n*sumval
    ginicoef=(2*gini)/giniden-(n+1)/n
    print(ginicoef)
        

data=pd.read_csv('adult.csv')
d=data.iloc[:,0]
gini_coefficient(d)
```
